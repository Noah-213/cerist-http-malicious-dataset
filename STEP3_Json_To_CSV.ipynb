```python
CLEANED_FLATTENED_DEDUP_DIR = "/content/drive/MyDrive/BINOME_WORK/STAGE_CERIST/DATA_FORMATING/STEP2_json_CLEANED_FLATTENED_DEDUP"
CSV_DIR = "/content/drive/MyDrive/BINOME_WORK/STAGE_CERIST/DATA_FORMATING/STEP3_CONVERTED_CSV"
os.makedirs(CSV_DIR, exist_ok=True)
```

```python
import os
import json
import csv

def json_to_csv(json_path, csv_path):
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    fieldnames = set()
    for entry in data:
        fieldnames.update(entry.keys())
    fieldnames = sorted(fieldnames)
    with open(csv_path, "w", encoding="utf-8", newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)
    print(f"[✔] {os.path.basename(json_path)} → {os.path.basename(csv_path)} ({len(data)} lignes)")

# Conversion de tous les fichiers JSON du dossier source vers le dossier destination
for filename in os.listdir(CLEANED_FLATTENED_DEDUP_DIR):
    if filename.endswith('.json'):
        json_path = os.path.join(CLEANED_FLATTENED_DEDUP_DIR, filename)
        csv_path  = os.path.join(CSV_DIR, filename.replace('.json', '.csv'))
        json_to_csv(json_path, csv_path)
```
